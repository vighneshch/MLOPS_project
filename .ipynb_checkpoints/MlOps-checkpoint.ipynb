```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from surprise.model_selection import cross_validate,train_test_split,GridSearchCV
from surprise import Dataset,SVD,Reader,accuracy
import mlflow

import os 
os.environ['GOOGLE_APPLICATION_CREDENTIALS']='D:\DataScience\Term 2\7.MLOPS\MLOPS final assignment\master-shore-411113-41df45dfb65e.json'
```

```python
ratings = pd.read_csv('D:\\DataScience\\Term 2\\7.MLOPS\\MLOPS final assignment\\MLOPS_project\\ml-latest-small\\ml-latest-small\\ratings.csv')
ratings.head()
```

```python
movie_meta = pd.read_csv('D:\\DataScience\\Term 2\\7.MLOPS\\MLOPS final assignment\\MLOPS_project\\ml-latest-small\\ml-latest-small\\movies.csv')
movie_meta.head()
```

```python
ratings_movie_title = pd.merge(left=ratings,right = movie_meta,how='left',left_on = 'movieId',right_on='movieId')
ratings_movie_title.drop(['timestamp','genres'],inplace=True,axis=1)
ratings_movie_title.head()
```

```python

movie_ratings_pivot = ratings_movie_title.pivot(index=['title','movieId'],columns='userId',values='rating').fillna(0)
movie_ratings_pivot.head()
```

```python
movie_ratings_pivot = movie_ratings_pivot.droplevel(level='movieId')
movie_ratings_pivot.head()
```

## Model Based

```python
reader = Reader(rating_scale=(0,5))
data = Dataset.load_from_df(ratings_movie_title[['userId','title','rating']],reader)
data.raw_ratings[:5]
```

```python
trainset,testset = train_test_split(data,test_size=0.2,random_state=42)
```

```python
#Model fit

model = SVD()
model.fit(trainset)
```

```python
#train_prediction = model.test(trainset)
test_prediction = model.test(testset)

accuracy.rmse(test_prediction)
```

```python
#Tuning algorithm

param_grid = {"n_epochs": [5,10,15,20,25,30,50,60], "lr_all": [0.002, 0.005,0.008], "reg_all": [0.4, 0.6]}
gs = GridSearchCV(SVD, param_grid, measures=["rmse", "mae"], cv=3)
gs.fit(data)
```

```python
print(gs.best_score)
```

```python
print(gs.best_params['rmse'])
```

```python
#SVD with the tuned parameteres

model_svd_tuned = SVD(n_epochs= 60, lr_all= 0.005, reg_all= 0.4)
model_svd_tuned.fit(trainset)
tuned_test_prediction = model_svd_tuned.test(testset)
accuracy.rmse(tuned_test_prediction)
```

```python
#mlflow
mlflow.set_tracking_uri('http://34.100.213.14:5000/')

exp_name = 'Movie_recommender_system'

#Creating exp_name in mlflow server
if mlflow.get_experiment_by_name(exp_name) is None:
    mlflow.create_experiment(exp_name)
else:
    mlflow.set_experiment(exp_name)

#creating run for SVD

with mlflow.start_run():
    mlflow.log_params(gs.best_params['rmse'])
    mlflow.log_metrics({'rmse':0.8826})
    mlflow.sklearn.log_model(model_svd_tuned,'SVD')
```
